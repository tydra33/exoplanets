# Coordinates

## Calculate Cartesian coordinates of host system
### How do spherical coordinates work?
![alt text](images/spherical_coords.png "Spherical Coordinates")

Spherical coordinates are made up of:
1. the radial distance `r` along the line connecting the point to a fixed point called the origin (`sy_dist`) <br>
the Euclidean distance from the origin O to P.

2. the polar angle `θ` between this radial line and a given polar axis (`ra`) <br>
angle between 'z-axis' and `r`.

3. the azimuthal angle `φ`, which is the angle of rotation of the radial line around the polar axis (`dec`) <br>
angle between 'x-axis' and `r`. Measured counterclockwise from X to Y.


### Equation for converting to Cartesian coordinate system
*Source: [Putting Your Stars in Their Places](https://fmwriters.com/Visionback/Issue14/wbputtingstars.htm)*  

$$x = distance \cdot \cos(dec) \cdot \cos(ra)$$
$$y = distance \cdot \cos(dec) \cdot \sin(ra)$$
$$z = distance \cdot \sin(dec)$$

We must convert degrees to radians in order to calculate Cartesian coords.

```python
planets_merged['dec_rad'] = np.radians(planets_merged['dec'])
planets_merged['ra_rad'] = np.radians(planets_merged['ra'])


planets_merged['st_x'] = planets_merged['sy_dist'] * np.cos(planets_merged['dec_rad']) * np.cos(planets_merged['ra_rad'])
planets_merged['st_y'] = planets_merged['sy_dist'] * np.cos(planets_merged['dec_rad']) * np.sin(planets_merged['ra_rad'])
planets_merged['st_z'] = planets_merged['sy_dist'] * np.sin(planets_merged['dec_rad'])
```

## Calculate Cartesian coordinates of planets
Each planet is positioned relative to its host star in 3D space. We use the same RA/Dec → Cartesian conversion logic as we did for the host, but instead of placing the planet exactly at the star’s coordinates, we offset it by its orbital distance to simulate a circular orbit around the star.

Steps Explained
1. Orbital Radius (pl_orbsmax)
This column gives the planet’s semi-major axis in AU (distance from the star).
It determines how far the planet is from its host.

2. Random Orbital Angle (theta)
We generate a random angle to place the planet somewhere along its orbital circle in the XY-plane.
This spreads planets around the star instead of stacking them on a single line.

3. Orbital Tilt (phi)
To simulate real orbital inclinations, we add a small tilt in the Z-axis.
This makes the orbits slightly 3D instead of perfectly flat.

To calculate the planetary offsets (relative to host), we perform the following (`r` equates to `orbital_radius`):
$$x_{\text{offset}} = r \cdot \cos(\theta) \cdot \cos(\phi)$$
$$y_{\text{offset}} = r \cdot \sin(\theta) \cdot \cos(\phi)$$
$$z_{\text{offset}} = r \cdot \sin(\phi)$$

Then we add the aforementioned offset to the host coords:
$$x_{\text{planet}} = x_{\text{host}} + x_{\text{offset}}$$
$$y_{\text{planet}} = y_{\text{host}} + y_{\text{offset}}$$
$$z_{\text{planet}} = z_{\text{host}} + z_{\text{offset}}$$

```python
# Random angle around star
theta = np.random.uniform(0, 2*np.pi, size=len(planets_merged))
phi = np.random.uniform(-0.1, 0.1, size=len(planets_merged))  # small tilt for variation

# Planet semi-major axes (AU) → scale for visualization
au_to_pc = 4.848e-6  # 1 AU in parsecs
scale = 1e4           # adjust multiplier for visibility in scene
orbital_radius = planets_merged['pl_orbsmax'] * au_to_pc * scale

planets_merged['pl_x'] = planets_merged['st_x'] + orbital_radius * np.cos(theta) * np.cos(phi)
planets_merged['pl_y'] = planets_merged['st_y'] + orbital_radius * np.sin(theta) * np.cos(phi)
planets_merged['pl_z'] = planets_merged['st_z'] + orbital_radius * np.sin(theta)
```

## Stellar classes
*Source: [Stellar classification](https://en.wikipedia.org/wiki/Stellar_classification)*  

Another element that will affect the visuals is **stellar classification**.

![alt text](images/stellar_classification.png "Stellar Classification")

In astronomy, stellar classification is the classification of stars based on their spectral characteristics. Electromagnetic radiation from the star is analyzed by splitting it with a prism or diffraction grating into a spectrum exhibiting the rainbow of colors interspersed with spectral lines. <br>

We already have a spectral type column, but since we're using means, we can recalculate them.

```python
# K = C + 273.15
class_ranges = {
    'O': (30273, 60000),
    'B': (9973, 30273),
    'A': (7473, 9973),
    'F': (5973, 7473),
    'G': (5173, 5973),
    'K': (3673, 5173),
    'M': (2373, 3673)
}

class_colors = {
    'O': ((155,176,255),(120,150,255)),
    'B': ((170,191,255),(150,170,255)),
    'A': ((202,215,255),(180,200,255)),
    'F': ((248,247,255),(255,240,230)),
    'G': ((255,244,234),(255,235,210)),
    'K': ((255,210,161),(255,180,130)),
    'M': ((255,204,111),(255,160,80))
}
```

```python
def classify_star(teff):
    if teff >= class_ranges['O'][0]:    # >30,000 °C
        return 'O'
    elif teff >= class_ranges['B'][0]:  # 9,700–30,000 °C
        return 'B'
    elif teff >= class_ranges['A'][0]:  # 7,200–9,700 °C
        return 'A'
    elif teff >= class_ranges['F'][0]:  # 5,700–7,200 °C
        return 'F'
    elif teff >= class_ranges['G'][0]:  # 4,900–5,700 °C
        return 'G'
    elif teff >= class_ranges['K'][0]:  # 3,400–4,900 °C
        return 'K'
    elif teff >= class_ranges['M'][0]:  # 2,100–3,400 °C
        return 'M'
    else:
        return 'Unknown'  # For extremely cool objects (brown dwarfs, etc.)
```

```python
def interpolate_color(start_rgb, end_rgb, factor):
    return tuple(int(start + (end - start) * factor) for start, end in zip(start_rgb, end_rgb))

def temp_to_class_color(teff, spectral_class):
    if spectral_class not in class_ranges:
        return (128,128,128)
    tmin, tmax = class_ranges[spectral_class]
    start_rgb, end_rgb = class_colors[spectral_class]
    factor = (teff - tmin) / (tmax - tmin)
    factor = max(0, min(1, factor))  # clamp
    return interpolate_color(start_rgb, end_rgb, factor)
```

```python
planets_merged['spectral_class'] = planets_merged['st_teff'].apply(classify_star)
planets_merged.groupby('spectral_class')['rowid'].count()
```

```python
planets_merged['st_color'] = planets_merged.apply(
    lambda row: temp_to_class_color(row['st_teff'], row['spectral_class']), axis=1
)
planets_merged
```
